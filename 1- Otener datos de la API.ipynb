```python
url = "https://api.twitter.com/2/tweets/search/recent"
```

```python
params = {
    'query': '#machinelearning -is:retweet', #Va a traer los que tienen el hashtag '#machinelearning' y el -is:retweet es para que no traigra retweets
    'tweet.fields':'created_at', #Treae la fecha de creacion
    'max_results':10 #Y que traiga 100 resultados maximo
}
```

```python
headers = {
    "Authorization": f"Bearer {bearer_token}",
    "User-Agent":"v2FullArchiveSearchPython"
}
```

```python
import requests
response = requests.get(url, headers=headers, params=params)
print(response)
# Generar excepción si la respuesta no es exitosa
if response.status_code != 200:
    raise Exception(response.status_code, response.text)
print(response.json())
```

```python
import pandas as pd
df = pd.json_normalize(response.json()['data'])
df
```
